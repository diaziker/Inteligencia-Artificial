Definimos función para cambiar un precio cómo string a float

```python
def value_to_float(x):
    """
    From K and M to float.
    
    """
    
    if(type(x)==str):
        x = x.replace('€', '')
        ret_val = 0.0

        if type(x) == float or type(x) == int:
            ret_val = x
        if 'K' in x:
            if len(x) > 1:
                ret_val = float(x.replace('K', ''))
            ret_val = ret_val *1000
        if 'M' in x:
            if len(x) > 1:
                ret_val = float(x.replace('M', ''))
            ret_val = ret_val * 1000000.0
        return ret_val
    return 0.0
```

Funcion para cambiar de peso cómo string a float

```python
def weight_to_float(x):
    """
    From weight to Float
    """
    weight = 0.0
    
    if(type(x)==str):
        lbs = x[:3]
        weight = float(lbs)
        return weight
    return 0.0
    
```

Función para sumar las expectativas en cada posicion de un jugador en string a float

```python
def pos_value_to_float(x):
    """
    From position with + to Float
    """
    pos = 0.0
    if(type(x)==str):
        first = x[:2]
        first = float(first)
        second = x[-1:]
        second = float(second)
        pos = first + second
        return pos
    return 0.0
    
```

Función para cambiar altura de string a float

```python
def height_to_float(x):
    """
    From height to Float
    """
    
    if(type(x)==str):
        first = x[0]
        first = float(first)
        second = x[2:]
        second = float(second)
        second = second/100
        height = first + second
        return height
    return 0.0
    
```

Comprobamos los jugadores que tienen *NaN* en la columna 'Release Clause', pero observamos que contienen muchos otros datos que pueden sernos util para el entranamiento por tanto, utilizamos la función que en caso de ser *NaN* le asigne el valor 0

```python
df.loc[df["Release Clause"] != df["Release Clause"]]
```

Aplicamos las funciones correspondientes de cada columna para que nos queden solo valores numéricos

```python
df["Value"] = df["Value"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)
df["Release Clause"] = df["Release Clause"].apply(value_to_float)
df["Weight"] = df["Weight"].apply(weight_to_float)
df["Height"] = df["Height"].apply(height_to_float)
df
```

Aplicamos la función posicón a float para cada una de las proyecciones por posición de cada jugador

```python
df["LS"] = df["LS"].apply(pos_value_to_float)
df["ST"] = df["ST"].apply(pos_value_to_float)
df["RS"] = df["RS"].apply(pos_value_to_float)
df["LW"] = df["LW"].apply(pos_value_to_float)
df["LF"] = df["LF"].apply(pos_value_to_float)
df["CF"] = df["CF"].apply(pos_value_to_float)
df["RF"] = df["RF"].apply(pos_value_to_float)
df["RW"] = df["RW"].apply(pos_value_to_float)
df["LAM"] = df["LAM"].apply(pos_value_to_float)
df["CAM"] = df["CAM"].apply(pos_value_to_float)
df["RAM"] = df["RAM"].apply(pos_value_to_float)
df["LM"] = df["LM"].apply(pos_value_to_float)
df["LCM"] = df["LCM"].apply(pos_value_to_float)
df["CM"] = df["CM"].apply(pos_value_to_float)
df["RCM"] = df["RCM"].apply(pos_value_to_float)
df["RM"] = df["RM"].apply(pos_value_to_float)
df["LWB"] = df["LWB"].apply(pos_value_to_float)
df["LDM"] = df["LDM"].apply(pos_value_to_float)
df["CDM"] = df["CDM"].apply(pos_value_to_float)
df["RDM"] = df["RDM"].apply(pos_value_to_float)
df["RWB"] = df["RWB"].apply(pos_value_to_float)
df["LB"] = df["LB"].apply(pos_value_to_float)
df["LCB"] = df["LCB"].apply(pos_value_to_float)
df["CB"] = df["CB"].apply(pos_value_to_float)
df["RCB"] = df["RCB"].apply(pos_value_to_float)
df["RB"] = df["RB"].apply(pos_value_to_float)

df
```

## Prediction

Finally we want to try to predict the **value**

```python
val = df.pop("Value")
```

At the same time we have the data we want to predict in a numeric format, the value information. So now we can use this data to train a model. But first we need to split the data to being able to known its true performance. Utilizamos un 20% tests y 80% predicción de entrenamiento

```python
X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.2, random_state=42)
```

```python
len(X_train)
```

Now we train a LinearRegression model.

```python
reg = linear_model.LinearRegression().fit(X_train, y_train)
```

Finally we get a metric $R^2$ for the regression, we use the implementation from <a href="https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html"> sickit-learn</a>.

```python
preds = reg.predict(X_test)
```

Tenemos la predicción del primer jugador

Test del primer jugador

```python
y_test[0]
```

```python
r2_score(preds, y_test)
```
